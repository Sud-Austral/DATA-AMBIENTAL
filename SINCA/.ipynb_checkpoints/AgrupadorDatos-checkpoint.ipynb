```python
salidafecha = []
fecha = df1["FECHA..YYMMDD."]
for i in fecha:
    d = i[-2:]
    m = i[-4:-2]
    y = int(i[-6:-4])+2000
    y = str(y)
    #if(len(y)==1):
    #    y = "200"+y
    print(d+"/"+m+"/"+y)
    salidafecha.append(datetime.strptime(d+"/"+m+"/"+y, '%d/%m/%Y'))
salidafecha
```

```python
df1["Fecha"]=salidafecha
df1
```

```python
#df1.drop(['FECHA..YYMMDD.'], axis = 1)
salida = df1.drop(['FECHA..YYMMDD.'], axis=1)
salida.columns
```
